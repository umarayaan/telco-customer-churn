## EDA 

```python
sns.countplot(x = 'Churn', data = df) #data is biased towards non churned customers
plt.title('Distribution of Churned vs Non Churned Customers')
```

```python
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.show() ## senior citizens are more likely to churn than non senior citizens
```

```python
sns.barplot(x=df.MonthlyCharges,y=df.Churn,data=df)## customers with higher monthly charges are more likely to churn than those with lower monthly charges
```

```python
sns.stripplot(x='Contract',y='MonthlyCharges',data=df) ## customers with month to month contract are more likely to have higher monthly charges than those with one year or two year contract
```
