## scaling 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x = new_df.drop('Churn', axis = 1)
y = new_df['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
std  = StandardScaler()
x_train = std.fit_transform(x_train)
x_test = std.transform(x_test)
```

## smote 

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_sm, y_train_sm = smote.fit_resample(x_train, y_train)
```

## models 

##  logistic regression 

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 
```

```python
lin_mod = LogisticRegression()
lin_mod
lin_mod.fit(x_train_sm, y_train_sm)
y_pred = lin_mod.predict(x_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

## decision tree classifier

```python
decis_tree = DecisionTreeClassifier()
decis_tree.fit(x_train_sm, y_train_sm)   
decis_tree_pred = decis_tree.predict(x_test)
print(classification_report(y_test, decis_tree_pred))
```

## svm (support vector machine)

```python
svc_model = SVC()
svc_model.fit(x_train_sm, y_train_sm) 
svc_model_pred = svc_model.predict(x_test)
print(classification_report(y_test, svc_model_pred))    
```

## random forest

```python
rand_forest = RandomForestClassifier()
rand_forest.fit(x_train_sm, y_train_sm) 
rand_forest_pred = rand_forest.predict(x_test)
print(classification_report(y_test, rand_forest_pred))
```

## ann 

```python
from sklearn.neural_network import MLPClassifier   
mlp = MLPClassifier(hidden_layer_sizes=(32, 16),
                activation='relu', solver='adam', max_iter=500)
mlp.fit(x_train_sm, y_train_sm)
mlp_pred = mlp.predict(x_test)
print(classification_report(y_test, mlp_pred))
```
